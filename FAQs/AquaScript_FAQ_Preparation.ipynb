## Aquascript API Documentation

## Movies API

### General

1. **What is the Aquascript Movies API?**
   - It's a JSON-based API offering movie metadata including genres, directors, download links, and more.

2. **What types of data does the Movies API provide?**
   - It provides movie titles, release year, genres, directors, and download links.

3. **Is the Movies API free to use?**
   - Yes, the Movies API is free for public use and can be accessed without any cost.

4. **What format is the data returned in?**
   - The data is returned in JSON format, which can be easily processed in JavaScript.

### Data Format

1. **What does moviesdata.json contain?**
   - It contains an array of movie genres and a list of movie objects with basic details like title, year, and director.

2. **What extra information does moviesdata++.json provide?**
   - It includes tags, director, release date, quality, size, direct download links, and poster images.

3. **Can I filter movies by genre using the Movies API?**
   - Yes, movies are categorized by genre in the moviesdata.json file, which allows filtering.

4. **How are movie images structured in the API?**
   - Images are provided with URLs, pointing to the movie posters, accessible in the image field of each movie object.

### Usage

1. **How can I fetch movie data using JavaScript?**
   - Use fetch('path/to/moviesdata.json').then(response => response.json()).then(data => { /* handle data */ });

2. **How can I fetch extended movie data using JavaScript?**
   - Use fetch('path/to/moviesdata++.json').then(response => response.json()).then(data => { /* handle extended data */ });

3. **How can I display the fetched movie data on a webpage?**
   - You can loop through the movie data in JavaScript and dynamically generate HTML elements to display each movie's details.

4. **How can I use JavaScript to create movie cards?**
   - Create a div for each movie with elements like title, year, genres, director, and poster, then append these to the main list container.

### Best Practices

1. **What are best practices for using the Movies API?**
   - Validate links, lazy load images, use semantic HTML, and handle API errors gracefully.

2. **How should I handle broken movie links?**
   - Always check the validity of external URLs (e.g., download links) and handle them with proper error messaging.

3. **What is lazy loading and how does it improve performance?**
   - Lazy loading involves loading images only when they are about to be visible on the screen, improving page load time.

4. **How can I handle errors gracefully while fetching movie data?**
   - Use `.catch()` to handle errors in the fetch request, and display a user-friendly message if data can't be loaded.

## Books API

### General

1. **What is the Aquascript Book API?**
   - It's a lightweight JSON RESTful API that provides metadata for books such as author, country, language, and publication year.

2. **How can I use the Books API in my app?**
   - You can use the Books API to display books, their metadata, and add functionality like sorting or filtering based on attributes like author or year.

### Data Format

1. **What data is returned for each book in the Books API?**
   - Each book object contains title, author, language, country, image link, Wikipedia link, pages, and publication year.

2. **How is the book cover image structured in the Books API?**
   - The image link field in each book object contains the URL of the book's cover image.

### Usage

1. **How do I fetch book data using JavaScript?**
   - Use fetch('https://wecoded-dev.github.io/Aquascript/api/books.json') and parse the JSON response.

2. **How can I display book data on a webpage?**
   - Loop through the book objects and create HTML elements (e.g., divs, images, and paragraphs) to display the data.

### Best Practices

1. **What are the best practices for using the Books API?**
   - Validate URLs, use lazy loading for images, structure content semantically, and add error handling.

2. **How should I handle broken book links?**
   - Always check if the external URLs (like Wikipedia) are valid before linking them in your app.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings

# Ignore warnings
warnings.filterwarnings('ignore')

# Make a copy to avoid modifying the original data too early
df_copy = df_combined.copy()

# Count the number of questions per category and subcategory
category_counts_before = df_copy['Category'].value_counts()
subcategory_counts_before = df_copy['Subcategory'].value_counts()

# Compute percentages
category_percent = category_counts_before / category_counts_before.sum() * 100
subcategory_percent = subcategory_counts_before / subcategory_counts_before.sum() * 100

# Threshold for "Other"
threshold_category = 4
threshold_subcategory = 4

# Identify small categories and subcategories
small_categories = category_percent[category_percent < threshold_category].index
small_subcategories = subcategory_percent[subcategory_percent < threshold_subcategory].index

# Save original small category/subcategory counts to dicts
other_categories_dict = category_counts_before[small_categories].to_dict()
other_subcategories_dict = subcategory_counts_before[small_subcategories].to_dict()

# Debug info BEFORE replacement
print(f"\nðŸ“Œ Threshold Logic Applied:")
print(f"  - Categories < {threshold_category}% grouped into 'Other'")
print(f"  - Subcategories < {threshold_subcategory}% grouped into 'Other'\n")
print("\n ðŸ—‚ Small Categories grouped into 'Other':\n\n", other_categories_dict)
print("\n ðŸ“ Small Subcategories grouped into 'Other':\n\n", other_subcategories_dict)

# Replace small ones with "Other"
df_copy['Category'] = df_copy['Category'].replace(small_categories, 'Other')
df_copy['Subcategory'] = df_copy['Subcategory'].replace(small_subcategories, 'Other')

# Recompute counts after replacement
category_question_count = df_copy['Category'].value_counts()
subcategory_question_count = df_copy['Subcategory'].value_counts()

# ------------------- Pie Chart: Categories -------------------
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

print("\n\n ========= Category Distribution (Excluding 'Other') ================== \n\n")

category_main = category_question_count[category_question_count.index != 'Other']
wedges, texts, autotexts = axes[0].pie(
    category_main,
    autopct='%1.1f%%',
    colors=sns.color_palette("pastel"),
    startangle=140,
    labels=category_main.index
)
for text in texts:
    text.set_horizontalalignment('center')

axes[0].set_title("Category Distribution (Excluding 'Other')")
axes[0].set_ylabel('')

print("\n\n ========= Detailed Breakdown of 'Other' Category ================== \n\n")

# Detailed pie chart for 'Other' components
if other_categories_dict:
    wedges2, texts2, autotexts2 = axes[1].pie(
        other_categories_dict.values(),
        labels=other_categories_dict.keys(),
        autopct='%1.1f%%',
        colors=sns.color_palette("pastel", len(other_categories_dict)),
        startangle=140
    )
    for text in texts2:
        text.set_rotation(90)
        text.set_horizontalalignment('center')
    axes[1].set_title("Detailed Breakdown of 'Other' Category")
else:
    axes[1].text(0, 0, "No Small Categories", ha='center', va='center', fontsize=14)

axes[1].set_ylabel('')
plt.tight_layout()
plt.show()

# ------------------- Bar Chart: Small Categories -------------------

print("\n\n ========= Bar Chart Detailed Breakdown of 'Other' Category ================== \n\n")

if other_categories_dict:
    # First separate bar chart
    plt.figure(figsize=(10, 6))
    plt.bar(
        x=list(other_categories_dict.keys())[:10],
        height=list(other_categories_dict.values())[:10],
        color=sns.color_palette("pastel", 10)
    )
    plt.title("Top 10 Small Categories (< Threshold)")
    plt.xlabel("Category")
    plt.ylabel("Number of Questions")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Second separate bar chart
    if len(other_categories_dict) > 10:
        plt.figure(figsize=(10, 6))
        plt.bar(
            x=list(other_categories_dict.keys())[10:],
            height=list(other_categories_dict.values())[10:],
            color=sns.color_palette("pastel", len(other_categories_dict) - 10)
        )
        plt.title("Remaining Small Categories")
        plt.xlabel("Category")
        plt.ylabel("Number of Questions")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

    # Full detailed bar chart for all small categories
    plt.figure(figsize=(10, 6))
    plt.bar(
        x=list(other_categories_dict.keys()),
        height=list(other_categories_dict.values()),
        color=sns.color_palette("pastel", len(other_categories_dict))
    )
    plt.title("All Small Categories (< Threshold)")
    plt.xlabel("Category")
    plt.ylabel("Number of Questions")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

# ------------------- Pie Chart: Subcategories -------------------
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

print("\n\n ========= Subcategory Distribution (Excluding 'Other') ================== \n\n")

subcategory_main = subcategory_question_count[subcategory_question_count.index != 'Other']
wedges3, texts3, autotexts3 = axes[0].pie(
    subcategory_main,
    autopct='%1.1f%%',
    colors=sns.color_palette("pastel"),
    startangle=140,
    labels=subcategory_main.index
)

axes[0].set_title("Subcategory Distribution (Excluding 'Other')")
axes[0].set_ylabel('')

print("\n\n ========= Detailed Breakdown of 'Other' Subcategory ================== \n\n")

if other_subcategories_dict:
    wedges4, texts4, autotexts4 = axes[1].pie(
        other_subcategories_dict.values(),
        labels=other_subcategories_dict.keys(),
        autopct='%1.1f%%',
        colors=sns.color_palette("pastel", len(other_subcategories_dict)),
        startangle=140
    )
    for text in texts4:
        text.set_rotation(90)
        text.set_horizontalalignment('center')
    axes[1].set_title("Detailed Breakdown of 'Other' Subcategory")
else:
    axes[1].text(0, 0, "No Small Subcategories", ha='center', va='center', fontsize=14)

axes[1].set_ylabel('')
plt.tight_layout()
plt.show()

# ------------------- Bar Chart: Small Subcategories -------------------

print("\n\n ========= Bar Chart of Detailed Breakdown of 'Other' Subcategory ================== \n\n")

if other_subcategories_dict:
    # First separate bar chart
    plt.figure(figsize=(10, 6))
    plt.bar(
        x=list(other_subcategories_dict.keys())[:10],
        height=list(other_subcategories_dict.values())[:10],
        color=sns.color_palette("pastel", 10)
    )
    plt.title("Top 10 Small Subcategories (< Threshold)")
    plt.xlabel("Subcategory")
    plt.ylabel("Number of Questions")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Second separate bar chart
    if len(other_subcategories_dict) > 10:
        plt.figure(figsize=(10, 6))
        plt.bar(
            x=list(other_subcategories_dict.keys())[10:],
            height=list(other_subcategories_dict.values())[10:],
            color=sns.color_palette("pastel", len(other_subcategories_dict) - 10)
        )
        plt.title("Remaining Small Subcategories")
        plt.xlabel("Subcategory")
        plt.ylabel("Number of Questions")
        plt.xticks(rotation=90)
        plt.tight_layout()
        plt.show()

    # Full detailed bar chart for all small subcategories
    plt.figure(figsize=(10, 6))
    plt.bar(
        x=list(other_subcategories_dict.keys()),
        height=list(other_subcategories_dict.values()),
        color=sns.color_palette("pastel", len(other_subcategories_dict))
    )
    plt.title("All Small Subcategories (< Threshold)")
    plt.xlabel("Subcategory")
    plt.ylabel("Number of Questions")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

### Word Count Histogram

```python
# Define word count function
def count_words(text):
    if isinstance(text, str):
        return len(text.split())
    return 0

# Word count processing
df_combined['Question_Word_Count'] = df_combined['Question'].apply(count_words)
df_combined['Answer_Word_Count'] = df_combined['Answer'].apply(count_words)
df_combined['Total_Word_Count'] = df_combined['Question_Word_Count'] + df_combined['Answer_Word_Count']

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot histogram for Question Word Count in the first subplot
sns.histplot(df_combined['Question_Word_Count'], bins=20, kde=True, color='skyblue', ax=axes[0], label='Question Word Count', alpha=0.7)
axes[0].set_title("Histogram of Question Word Count", fontsize=14)
axes[0].set_xlabel("Word Count")
axes[0].set_ylabel("Frequency")
axes[0].legend()

# Plot histogram for Answer Word Count in the second subplot
sns.histplot(df_combined['Answer_Word_Count'], bins=20, kde=True, color='lightgreen', ax=axes[1], label='Answer Word Count', alpha=0.7)
axes[1].set_title("Histogram of Answer Word Count", fontsize=14)
axes[1].set_xlabel("Word Count")
axes[1].set_ylabel("Frequency")
axes[1].legend()

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
